# **Checkpoints - Part 1**

## **Checkpoint 1: Data Cleaning 1**

```python
# importing the rounds2 csv file to a dataframe

rounds2 = pd.read_csv("rounds2.csv",encoding='ISO-8859-1')
rounds2
```

```python
#Removing special characters in column company_permalink 

rounds2['company_permalink']=rounds2.company_permalink.str.encode('ISO-8859-1').str.decode('ascii', 'ignore')
```

```python
# importing the companies text file to a dataframe

companies= pd.read_csv("companies.txt",sep='\t',encoding='ISO-8859-1')
companies
```

```python
#Removing special characters in columns permalink and name

companies['permalink']=companies.permalink.str.encode('ISO-8859-1').str.decode('ascii', 'ignore')
companies['name']=companies.name.str.encode('ISO-8859-1').str.decode('ascii', 'ignore')
```

```python
# unique companies in rounds2
rounds2.company_permalink.str.lower().nunique(dropna=True)   # dropna=True will exclude null values
```

```python
# unique companies in companies
companies.permalink.str.lower().nunique(dropna=True)
```

```python
# converting the values in company_permalink to lower case
rounds2['company_permalink']=rounds2.company_permalink.str.lower()

# converting the values in column permalink to lower case
companies['permalink']=companies.permalink.str.lower()
```

```python
# checking if there are any companies in the rounds2 dataframe are not present in companies dataframe

len(set(rounds2['company_permalink'])-set(companies['permalink']))
```

#### As it is Zero, we can confirm that there are no companies exists in the rounds2 dataframe that are not present in companies dataframe

```python
# another way of checking if there are any companies in the rounds2 dataframe are not present in companies dataframe

(rounds2[rounds2.company_permalink.isin(companies.permalink)]).shape[0] # gives no. of observations i.e rows
```

```python
rounds2.shape[0]
```

#### As no. of observations is same, we can confirm that there are no companies exists in the rounds2 dataframe that are not present in companies dataframe

```python
# creating master data frame by merging rounds2 and companies

master_frame=pd.merge(rounds2,companies,left_on='company_permalink',right_on='permalink',how='inner')
```

```python
master_frame
```

```python
# no. of observations in merged data frame i.e. master_frame

master_frame.shape[0]
```

```python
# Dropping unnecessary columns which are not used for analysis

master_frame=master_frame.drop(['funded_at',
                                'homepage_url',
                                'funding_round_permalink',
                                'permalink',
                                'status',
                                'founded_at',
                                'state_code',
                                'city',
                                'region'
                                ]
                                 ,axis=1)
```

```python
# Dropping the rows having null values in column raised_amount_usd

master_frame= master_frame[~np.isnan(master_frame.raised_amount_usd)]
```

```python
# There are few rows having NaN values in country_code but not deleting them as there are raised_amount_usd values exists in 
#those rows. so, we are not cleaning NaN values in country_code by assuming above reason


# Not deleting rows having NaN values in funding_round_code as the funding type and amount related to it will be lost.

# Not deleting rows having blank or NaN values in category_list at present assuming it will be used in later part of analysis with 
# mapping file
```

```python
# raised_amount_usd will be measured in million $ from now.

# converting raised_amount_usd from $ to million $

master_frame.raised_amount_usd=master_frame.raised_amount_usd/1000000
```

## **Checkpoint 4: Sector Analysis 1**

```python
# Extracting the primary sector of each category list from the category_list column

master_frame['primary_sector']=master_frame.category_list.str.split('|',expand=True)[0]
```

```python
# importing the mapping csv file to a dataframe

mapping=pd.read_csv('mapping.csv')
```

```python
# removing the rows having null values in category_list as blanks not considered among 8 main sectors.

mapping=mapping[~pd.isnull(mapping.category_list)]
mapping
```

```python
# handling some values in category_list column of mapping dataframe

# replaces '0' in the string with 'na' only if the below pattern matches with any of string.

mapping['category_list']=mapping.category_list.str.replace('(?<=\w)*0(\s)*(?=\w)','na')   
```

```python
# mapping each primary sector to one of the eight main sectors into a new column main_sector

mapping['main_sector']=mapping.apply(lambda x: x.index[x.astype(bool)].tolist()[1],axis=1) 
```

```python
# Merging the master_frame and mapping dataframes 

master_frame=pd.merge(master_frame,mapping,left_on='primary_sector',right_on='category_list',how='inner')
```

```python
# dropping unneccesary columns in master_frame

master_frame=master_frame.drop(['category_list_y','Automotive & Sports','Blanks','Cleantech / Semiconductors',
                                 'Entertainment','Health','Manufacturing','News, Search and Messaging','Others',
                                  'Social, Finance, Analytics, Advertising'
                                 ],axis=1)
```

```python
# renaming the column as name changed while merging mapping and master_frame dataframes


master_frame.rename(columns={'category_list_x':'category_list'},inplace=True)
master_frame
```

## **Checkpoint 5: Sector Analysis 2**

```python
# Creating new data frame D1 for country1: USA

D1= master_frame[(master_frame.country_code=='USA') & 
                  (master_frame.funding_round_type=='venture') &
                  ((master_frame.raised_amount_usd>=5) & (master_frame.raised_amount_usd<=15))
                 ]
```

```python
# Total number (or count) of investments for each main sector in new column

D1['count_investments_per_mainSector']=D1.groupby('main_sector').raised_amount_usd.transform('count')
```

```python
# Total amount invested in each main sector in a separate column

D1['Total_amount_per_mainSector']= D1.groupby('main_sector')["raised_amount_usd"].transform('sum')
D1
```
